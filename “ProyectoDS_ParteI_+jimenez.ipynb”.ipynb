```python
# Importar las librerías necesarias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Cargar el archivo CSV con codificación 'latin1' y separador ';'
df = pd.read_csv('defuncionesviales.csv', sep=';', encoding='latin1')

# Verificar si hay valores NaN
print("Valores NaN antes de la imputación:\n", df.isnull().sum())

# Opciones para manejar valores NaN: 
# Puedes imputarlos con la media, moda, o eliminarlos.
# En este ejemplo, se imputarán con la media (para valores numéricos) o la moda (para valores categóricos).

# Imputar valores NaN numéricos con la media y categóricos con la moda
for column in df.columns:
    if df[column].dtype == np.number:
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)

# Verificar que no haya más valores NaN
print("Valores NaN después de la imputación:\n", df.isnull().sum())

# Codificar las variables categóricas con LabelEncoder (si es necesario)
le = LabelEncoder()

# Aplicar LabelEncoder a todas las columnas categóricas
for col in df.select_dtypes(include=['object']).columns:
    df[col] = le.fit_transform(df[col])

# Separar las características (X) y la variable objetivo (y)
X = df.drop('clase victima', axis=1)  # Características
y = df['clase victima']  # Variable objetivo

# Dividir el dataset en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar los modelos
modelos = {
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    "Logistic Regression": LogisticRegression(max_iter=1000)
}

# Entrenar y evaluar cada modelo
resultados = {}
for nombre, modelo in modelos.items():
    print(f"\nEntrenando y evaluando el modelo: {nombre}")
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    
    # Evaluar las métricas de cada modelo
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    
    resultados[nombre] = {
        'accuracy': accuracy,
        'report': report
    }

# Imprimir los resultados comparativos
for nombre, resultado in resultados.items():
    print(f"\nModelo: {nombre}")
    print(f"Exactitud (Accuracy): {resultado['accuracy']}")
    print("Reporte de Clasificación:")
    print(resultado['report'])
```
